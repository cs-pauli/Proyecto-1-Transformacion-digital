```python
import re
import os
import xml.etree.ElementTree as ET
import os

tree = ET.parse('data/data_xml.xml')

root = tree.getroot()

xml_list = []
for child in root:
    xml_dict = {}
    for subchild in child:
        xml_dict[subchild.tag] = subchild.text
    xml_list.append(xml_dict) # Es una lista que contiene diccionarios
```

```python
# Explicacin tipos de datos xml_list:
# Vemos que es una lista de diccionarios, representando los registros de una tabla con varias columnas
# El diccionario se complone de: Keys (nombre de la columna) y Values (nombre del registro)
#En algunos diccionarios, vemos que estan con saltos de lnea entre los tems.
```

```python
# Cambiar la key gnero de int a str
```

```python
def limpieza_xml(archivo):

       for y in range(len(xml_list)):
                     xml_list[y]['index_xml'] = xml_list[y].pop('level_0')
                     xml_list[y]['index_sql'] = xml_list[y].pop('index')

       for i in xml_list:
              for k, v in i.items(): #por cada key y value de cada diccionario:
                     if k == "gender":       # si la key es = a gender (para que nos tome solo esa columna)
                            if v == "0":                     # si el valor es "0":
                                   i[k] = "Man"            # que vaya a la key de ese diccionario y cambie el value a "Man"
                            elif v == "1":                      # lo mismo para el resto
                                   i[k] = "Woman"
                            elif v == "2":
                                   i[k] = "Nonbinary"
                            elif v == "3":
                                   i[k] = "Prefer not to say"
                            elif v == "4":
                                   i[k] = "Prefer to self-describe"
       return archivo
```

```python
limpieza_xml(xml_list)
```

```python
def insertar_datos(query, datos, contrasea):

    import mysql.connector

    cnx = mysql.connector.connect(user='root', password=contrasea,
                                host='127.0.0.1', database= 'project1')


    mycursor = cnx.cursor()
    sql = query
    val = datos

    try: 
        mycursor.executemany(sql, val)
        cnx.commit()
        print(mycursor.rowcount, "registro/s insertado/s.")

    except mysql.connector.Error as err:
        print(err)
        print("Error Code:", err.errno)
        print("SQLSTATE", err.sqlstate)
        print("Message", err.msg)
```
